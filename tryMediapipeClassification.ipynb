```python
import mediapipe as mp
import cv2
import numpy as np
import uuid
import os
import csv
import pandas as pd 
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

```python
cap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5) as hands: 
    while cap.isOpened():
        ret, image = cap.read()
        
        # BGR 2 RGB
        image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
        
        # Set flag
        image.flags.writeable = False
        
        # Detections
        results = hands.process(image)
        
        # Set flag to true
        image.flags.writeable = True
        
        # RGB 2 BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        
        # Rendering results
        if results.multi_hand_landmarks:
            for hand in (results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS)
            
        # Save our image    
        
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            
            break

cap.release()
cv2.destroyAllWindows()
```

```python
#(TRAIN)
#Read Collected Data and process it
import pandas as pd 
from sklearn.model_selection import train_test_split
```

```python
#Read 'cords_new.csv'
df = pd.read_csv('cords_new.csv')
```

```python
df.head()
```

```python
# Features and Trget Variables
x = df.drop('class', axis=1) #feature
y= df['class'] #target
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.3, random_state=1234)
```

```python
y_train
```

```python
#CLASSIFICATION
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
#Algoithms for training
pipelines = {
    'lr': make_pipeline(StandardScaler(), LogisticRegression()),
    'rc': make_pipeline(StandardScaler(), RidgeClassifier()),
    'rf': make_pipeline(StandardScaler(), RandomForestClassifier()),
    'gb': make_pipeline(StandardScaler(), GradientBoostingClassifier()),
}
```

```python
fit_models ={}
for algo,pipeline in pipelines.items():
    model = pipeline.fit(x_train,y_train)
    fit_models[algo] = model
```

```python
fit_models
```

```python
#Prediction
fit_models['rf'].predict(x_test)
```

```python
#serialize Model
from sklearn.metrics import accuracy_score
import pickle
```

```python
#Evaluate Model
for algo, model in fit_models.items():
    pred = model.predict(x_test)
    print(algo, accuracy_score(y_test, pred))
```

```python
with open('hand_tracking.pkl', 'wb') as f:
    pickle.dump(fit_models['rf'], f)
```
